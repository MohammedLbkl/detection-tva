```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image
import torch

# Charger l'image
img = Image.open("../tmp/table_crop_11.png").convert("RGB")

# Charger le modèle trOCR
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-large-printed")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-large-printed")

# Préparer l'image pour le modèle
pixel_values = processor(images=img, return_tensors="pt").pixel_values

# Générer le texte
generated_ids = model.generate(pixel_values)
texte = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
```

```python
print(texte)
```

```python
import pytesseract
from PIL import Image
from pathlib import Path
import os
import re

folder_tmp = Path("../tmp")
liste_txt = []

files = list(folder_tmp.glob("*.png"))

files_sorted = sorted(
    files,
    key=lambda x: (
        int(re.search(r'table_crop_(\d+)', x.name).group(1))
    )
)
        
for img_path in files_sorted:
    
    img = Image.open(img_path)
    texte = pytesseract.image_to_string(img)
    liste_txt.append(texte)
    os.remove(img_path)
    
texte_final = "\n------------------------------------------------------------------\n\n".join(liste_txt)

print(texte_final)
```
